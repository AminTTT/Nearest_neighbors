```python
#                 .
#            .           
#     
# (2)                :   (0)-     (1)-    
#                 : https://www.geeksforgeeks.org/3d-scatter-plotting-in-python-using-matplotlib/
import matplotlib.pyplot as plt

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

x = [4, 5, 10, 4, 3, 11, 14 , 8, 10, 12]     #   (Tumor diameter)
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]         #  (Age of the patient)
z = [ 13, 3, 5, 8, 15, 13, 5, 6, 20, 12]   #        (The distance of the tumor from the vital organs of the body)
classes = [1, 0, 2, 2, 0, 1, 0, 2, 1, 0]
ax.scatter3D(x, y, z, c=classes)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("simple 3D scatter plot")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
data = list(zip(x,y,z))
print(data)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(data, classes)
```

```python
new_x = 10
new_y = 18
new_z = 15
new_point = [(new_x, new_y, new_z)]
prediction = knn.predict(new_point)
#          
print('class of new_point=',prediction)
```

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
ax.scatter3D(x+ [new_x], y+ [new_y], z+[new_z], c=classes+ [prediction[0]])
#  prediction[0]] :                       (  ): 
#                     1  
ax.text3D(x=new_x-1.7, y=new_y-0.7, z=new_z-4,   s=f"new point, class: {prediction[0]}")
plt.show()
```
